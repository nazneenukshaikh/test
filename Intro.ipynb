<div style="font-size:30px;font-weight:800;line-height:1.1;color:#000000;border-top:4px solid #EC0016;padding-top:16px;">What we check</div>
<div style="font-size:15px;color:#3a3a3a;margin-top:12px;max-width:840px;line-height:1.55;">Our checks line up with the standard data quality dimensions, so the coverage is easy to reason about: from whether the data is there, to whether it follows our business rules.</div>

<div style="display:flex;flex-wrap:wrap;gap:16px;margin-top:24px;">
<div style="flex:1 1 44%;min-width:300px;border:1px solid #E0E0E0;border-top:4px solid #EC0016;padding:18px 18px 16px;box-sizing:border-box;">
<div style="display:flex;justify-content:space-between;align-items:baseline;gap:10px;">
<div style="font-size:16px;font-weight:800;color:#000000;">Completeness</div>
<div style="font-size:12px;color:#EC0016;font-weight:700;white-space:nowrap;">Is the data there?</div></div>
<div style="font-size:13px;color:#3a3a3a;margin-top:11px;line-height:1.6;">Key fields are populated, identifiers are not blank, and numbers are real values rather than NaN or infinity.</div></div>
<div style="flex:1 1 44%;min-width:300px;border:1px solid #E0E0E0;border-top:4px solid #EC0016;padding:18px 18px 16px;box-sizing:border-box;">
<div style="display:flex;justify-content:space-between;align-items:baseline;gap:10px;">
<div style="font-size:16px;font-weight:800;color:#000000;">Validity</div>
<div style="font-size:12px;color:#EC0016;font-weight:700;white-space:nowrap;">Does it look right?</div></div>
<div style="font-size:13px;color:#3a3a3a;margin-top:11px;line-height:1.6;">Currencies, venues and statuses come from known lists. Amounts sit in sensible ranges. ISIN, LEI and email formats are checked, JSON and tag lists are well formed, and dates parse and are not in the future.</div></div>
<div style="flex:1 1 44%;min-width:300px;border:1px solid #E0E0E0;border-top:4px solid #EC0016;padding:18px 18px 16px;box-sizing:border-box;">
<div style="display:flex;justify-content:space-between;align-items:baseline;gap:10px;">
<div style="font-size:16px;font-weight:800;color:#000000;">Uniqueness</div>
<div style="font-size:12px;color:#EC0016;font-weight:700;white-space:nowrap;">Are we double-counting?</div></div>
<div style="font-size:13px;color:#3a3a3a;margin-top:11px;line-height:1.6;">No duplicate trades, and no duplicate client and date bookings. For large data we also run fast approximate checks.</div></div>
<div style="flex:1 1 44%;min-width:300px;border:1px solid #E0E0E0;border-top:4px solid #EC0016;padding:18px 18px 16px;box-sizing:border-box;">
<div style="display:flex;justify-content:space-between;align-items:baseline;gap:10px;">
<div style="font-size:16px;font-weight:800;color:#000000;">Consistency and business rules</div>
<div style="font-size:12px;color:#EC0016;font-weight:700;white-space:nowrap;">Do the fields agree?</div></div>
<div style="font-size:13px;color:#3a3a3a;margin-top:11px;line-height:1.6;">Net equals gross minus fees and tax, within tolerance. Settlement is on or after the trade date. If a trade is settled, it has a settlement date. Buy and sell match the sign of the cash flow.</div></div>
</div>

<div style="flex:1;min-width:180px;font-size:13px;line-height:1.5;"><b style="color:#EC0016;">Statistical and outliers.</b><br>We flag unusual values, for example beyond three standard deviations or above the 99th percentile.</div>
